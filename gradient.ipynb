```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function y = (x + 3)^2
def func(x):
    return (x + 3)**2

# Gradient of the function: derivative of y = (x + 3)^2 is 2(x + 3)
g = lambda x: 2 * (x + 3)

# Initialize variables
x = 2             # Starting point (x = 2)
p = 0.000001      # Precision for convergence
p_s = 1           # Previous step size (initial large value)
ite = 1           # Iteration counter
l_r = 0.01        # Learning rate
max_iter = 10000  # Maximum number of iterations
error = []        # To store the error history for visualization

# Gradient Descent loop
while p_s > p and ite < max_iter:
    y = x
    x = x - l_r * g(x)  # Update x using the gradient descent rule
    p_s = abs(y - x)    # Step size for convergence check
    ite += 1
    error.append(x)     # Append current value of x to error history

# Print the local minimum
print("Minimum value of x:", x)

# Visualization of the function and the gradient descent steps
x_vals = np.linspace(-10, 10)  # Generate values for x
y_vals = func(x_vals) # Calculate y for each x

# Plot the function y = (x + 3)^2
plt.plot(x_vals, y_vals, label="y = (x + 3)^2", color='b')

# Plot the points visited during gradient descent (red dots)
# plt.scatter(error, func(np.array(error)), color='r', label="Gradient Descent Steps")

# Labels and title
plt.title('Gradient Descent Optimization to Find Local Minima')
plt.xlabel('x')
plt.ylabel('y')

# Adding a legend to describe the plot
plt.legend()

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```
